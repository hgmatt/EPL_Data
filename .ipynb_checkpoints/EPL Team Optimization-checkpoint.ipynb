```python
#loading libraries
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import re
import StringIO
import pprint
import sys
%matplotlib inline
matplotlib.rc('font', size=18)
matplotlib.rc('figure', figsize=(12, 4))
```

```python
position_title = ['Goalkeeper', 'Defender', 'Forward', 'Midfielder']
points = {}
for p in all_data:
    if p['total_points'] >0:
        points.setdefault(p['element_type'],[]).append(p['total_points'])

def poshist(axis, position):
    axis.hist(points[position])
    
    axis.set_title(position_title[position-1])
    return axis

fig, (ax0, ax1, ax2, ax3) = plt.subplots(ncols=4, sharey=True, figsize=(18,4))
poshist(ax0, 1)
poshist(ax1, 2)
poshist(ax2, 3)
poshist(ax3, 4)
fig.show()
```

```python
ppg = {}
for p in all_data:
    if p['points_per_game'] >0:
        ppg.setdefault(p['element_type'],[]).append(float(p['points_per_game']))

def ppghist(axis, position):
    axis.hist(ppg[position])
    axis.set_title(position_title[position-1])
    return axis

fig, (ax0, ax1, ax2, ax3) = plt.subplots(ncols=4, sharey=True, figsize=(18,4))
ppghist(ax0, 1)
ppghist(ax1, 2)
ppghist(ax2, 3)
ppghist(ax3, 4)
fig.show()
```

```python
cost = {}
for p in all_data:
    cost.setdefault(p['element_type'],[]).append(float(p['now_cost']))

def costhist(axis, position):
    axis.hist(cost[position])
    axis.set_title(position_title[position-1])
    return axis

fig, (ax0, ax1, ax2, ax3) = plt.subplots(ncols=4, sharey=True, figsize=(18,4))
costhist(ax0, 1)
costhist(ax1, 2)
costhist(ax2, 3)
costhist(ax3, 4)
fig.show()
```

```python
cppg ={}
for p in all_data:
    if p['points_per_game'] >0:
        cppg.setdefault(p['element_type'],[]).append(10*float(p['points_per_game'])/float(p['now_cost']))

def cppghist(axis, position):
    axis.hist(cppg[position])
    axis.set_title(position_title[position-1])
    return axis

fig, (ax0, ax1, ax2, ax3) = plt.subplots(ncols=4, sharey=True, figsize=(18,4))
cppghist(ax0, 1)
cppghist(ax1, 2)
cppghist(ax2, 3)
cppghist(ax3, 4)
fig.show()
```
